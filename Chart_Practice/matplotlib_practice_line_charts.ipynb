```python
# Import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (method 1)
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()  (method 2)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()  (method 3)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()  (method 4)
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot  (method 5)
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# MATLAB Approach
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set the y limit between Jan and Jun.
#plt.xlim("Jan", "June") # also works - plt.xlim(0, 5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
# Change figure size # also works - plt.subplots(figsize=(20, 10))
#plt.figure(figsize=(20, 10))
```

```python
# SKILL DRILL Object-oriented approach
# Create the plot.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

# Error Bars

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Error bar function
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# Add cap to error bars
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#Object-Oriented Approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```
