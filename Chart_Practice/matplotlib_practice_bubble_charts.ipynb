```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Regular scatter plot
plt.scatter(x_axis, y_axis)
```

```python
# Bubble plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Bubble plot
plt.scatter(x_axis, y_axis, s=y_axis)
# for loop to iterate through all data and multiply by 3 to make bubble chart points larger
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Built in way to do the same as above chart
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

# Bubble Chart  Using Object-Oriented Approach

```python
# Create plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# SKILL DRILL
# Create plot
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue",s = [i * 5 for i in y_axis], alpha=.2, edgecolors='black', 
           linewidths = 2, label='Boston')
# Add the legend.
ax.legend()
# Create labels for the x and y axes.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Create a title.
ax.set_title("PyBer Fare by Month")
# Set the x limit between 0 and 45.
ax.set_xlim(0, 45)
plt.gca().invert_yaxis()
```
